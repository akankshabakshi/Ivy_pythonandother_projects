```python
#2. How many voyagers paying greater than mean fare are female ?
```

```python
titanic[(titanic['fare']>titanic['fare'].mean()) & (titanic['sex'] == "female")] #154 rows
```

```python
#3. How many male voyagers survived ?
```

```python
titanic.head(1)
```

```python
titanic[(titanic['sex'] == "male") & (titanic['survived'] == 1)] #161 rows
```

```python
#4. How many female voyagers didn't survive ?
```

```python
titanic[(titanic["sex"]== "female") & (titanic["survived"] == 0)] #127 rows
```

```python
#5. How many females survived ?
```

```python
titanic[(titanic["sex"]== "female") & (titanic["survived"] == 1)] #339 rows
```
