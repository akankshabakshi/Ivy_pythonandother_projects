```python
#Pivot tables
```

```python
df.pivot_table(index = "neighbourhood_group",
                columns = "room_type", 
                values = "price",
                aggfunc = "mean")
```

```python
#1. For each neighbourhood_group and neighbourhood by room type what is the average minimum nights offered by the hotels
```

```python
df.pivot_table(index = ["neighbourhood_group","neighbourhood"],
                columns = "room_type", 
                values = "minimum_nights",
                aggfunc = "mean")
```

```python
#2. Which neighbourhood has the minimum price for a room type
```

```python
df.pivot_table(index = ["neighbourhood"],
                columns = "room_type", 
                values = "price",
                aggfunc = "mean").sort_values(by = ["Shared room","Private room"])
```
